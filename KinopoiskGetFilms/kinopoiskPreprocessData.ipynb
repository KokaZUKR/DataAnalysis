```python
#some features can be grouped into classes:
# - finance data: 'budget', 'grossRU', 'grossUSA', 'grossWorld'
# - rating data: 'rating', 'ratingAgeLimits', 'ratingAwait', 'ratingAwaitCount', 'ratingFilmCritics',
#       'ratingFilmCriticsVoteCount', 'ratingGoodReview', 'ratingGoodReviewVoteCount', 
#       'ratingIMDb', 'ratingIMDbVoteCount',
#       'ratingMPAA', 'ratingVoteCount', 'reviewsCount'
# - logical data: 'hasSeance', 'is3D', 'isHasSequelsAndPrequelsFilms', 'isHasSimilarFilms', 'isIMAX'

financialFeatures = ['budget', 'grossRU', 'grossUSA', 'grossWorld']
ratingData = ['rating', 'ratingAgeLimits', 'ratingAwait', 'ratingAwaitCount', 'ratingFilmCritics',
              'ratingFilmCriticsVoteCount', 'ratingGoodReview', 'ratingGoodReviewVoteCount', 
              'ratingIMDb', 'ratingIMDbVoteCount',
              'ratingMPAA', 'ratingVoteCount', 'reviewsCount']
logicalFeatures = ['hasSeance', 'is3D', 'isHasSequelsAndPrequelsFilms',
                   'isHasSimilarFilms', 'isIMAX']
dateFeatures = ['premiereBluRay', 'premiereDVD', 'premiereRU', 'premiereWorld', 'rating']
```

```python
def makeCat(column):
    return column.astype("category")

films[['hasSeance', 'is3D', 'isHasSequelsAndPrequelsFilms','isHasSimilarFilms', 'isIMAX']] = films[['hasSeance', 'is3D', 'isHasSequelsAndPrequelsFilms','isHasSimilarFilms', 'isIMAX']].apply(makeCat,
                                                                                                axis=0)
#films['hasSeance'] = films['hasSeance'].astype("category")
```

```python
for feature in ['webURL','videoURL','posterURL','FilmID','filmID']:
    del films[feature]
```

```python
#convert date features to datetime
for feature in dateFeatures:
    films[feature] = pd.to_datetime(films[feature], errors='coerce')
```
