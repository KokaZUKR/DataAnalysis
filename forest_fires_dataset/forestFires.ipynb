```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.stats.api as sm
%matplotlib inline
%pylab inline

import warnings
warnings.filterwarnings('ignore')
```

```python
forest = pd.read_csv('./forestfires.csv', header = 0)
forest = forest.sample(frac=1)
```

```python
forest = forest[['area','RH','wind','rain','temp']]
```

```python
forest.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(forest.ix[:, forest.columns != 'area'], 
                                                    forest['area'], 
                                                    test_size=0.25)
```

```python
def jitter(arr):
    return arr + np.random.uniform(low=-0.35, high=0.35, size=len(arr))

pyplot.figure(figsize(16, 36))
for i in range (1, 5):
    pyplot.subplot(6, 2, i)
    pyplot.scatter(jitter(forest['area']), np.log10(forest.ix[:, i]), edgecolors="black")
    pyplot.xlabel('area', fontsize=14)
    pyplot.ylabel(str(forest.columns[i]), fontsize=14)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
np.sqrt(mean_squared_error(lm.predict(X_train), y_train))
```

```python
np.sqrt(mean_squared_error(lm.predict(X_test), y_test))
```

```python
plt.figure(figsize(16,7))
plt.subplot(121)
pyplot.scatter(np.log10(y_train), lm.predict(X_train), color="red", alpha=0.1)
pyplot.xlim(0,3)
pyplot.ylim(0,40)
plot(range(11), color='black')
grid()
pyplot.title('Train set', fontsize=20)
pyplot.xlabel('Quality', fontsize=14)
pyplot.ylabel('Estimated quality', fontsize=14)

plt.subplot(122)
pyplot.scatter(np.log10(y_test), lm.predict(X_test), color="red", alpha=0.1)
pyplot.xlim(0,3)
pyplot.ylim(0,40)
plot(range(11), color='black')
grid()
pyplot.title('Test set', fontsize=20)
pyplot.xlabel('Quality', fontsize=14)
pyplot.ylabel('Estimated quality', fontsize=14)
```

```python
lm.score(X_test, y_test)
```

Случайный лес

```python
rf = RandomForestRegressor(n_estimators=100, min_samples_leaf=2)
```

```python
rf.fit(X_train, y_train)
```

```python
sqrt(mean_squared_error(rf.predict(X_train), y_train))
```

```python
sqrt(mean_squared_error(rf.predict(X_test), y_test))
```

```python
plt.figure(figsize(16,7))
plt.subplot(121)
pyplot.scatter(y_train, lm.predict(X_train), color="red", alpha=0.1)
pyplot.xlim(0,70)
pyplot.ylim(0,40)
plot(range(100), color='black')
grid()
pyplot.title('Train set', fontsize=20)
pyplot.xlabel('Quality', fontsize=14)
pyplot.ylabel('Estimated quality', fontsize=14)

plt.subplot(122)
pyplot.scatter(y_test, lm.predict(X_test), color="red", alpha=0.1)
pyplot.xlim(0,100)
pyplot.ylim(0,40)
plot(range(70), color='black')
grid()
pyplot.title('Test set', fontsize=20)
pyplot.xlabel('Quality', fontsize=14)
pyplot.ylabel('Estimated quality', fontsize=14)
```

```python
rf.score(X_test, y_test)
```

```python
tmeans = sm.CompareMeans(sm.DescrStatsW(abs(y_test - lm.predict(X_test))), 
                         sm.DescrStatsW(abs(y_test - rf.predict(X_test))))

tmeans.ttest_ind(alternative='two-sided', usevar='pooled', value=0)[1]
```

```python
tmeans.tconfint_diff(alpha=0.05, alternative='two-sided', usevar='pooled')
```
