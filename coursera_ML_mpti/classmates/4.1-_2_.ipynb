```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

## Распределение хи-квадрат $\chi^2_k$ с $k$ степенями свободы

```python
k = 3
# Генерация выборки
rv = sts.chi2(k)
sample = rv.rvs(1000)
# Постройка гистограммы 
plt.hist(sample, normed=True, label='EHyst')
plt.ylabel('number of samples, %')
plt.xlabel('$x$')

# Постройка графика теоретической плотности
x = np.linspace(0,20,1000)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='Theoretical PDF', color='red')
plt.legend(loc='upper right')
plt.show()
```

### Выборка $n=5$

```python
k = 3.0
# Генерация параметров распределения
rv = sts.chi2(k)
n = 5.0
r = np.zeros(1000) #Массив под значения выборочных средних

# Получение массива выборочных средних
for i in range(1000):
    # Генерация выборки на i-м шаге и подсчет ее выборочного среднего
    sample = rv.rvs(n) 
    for j in sample:
        r[i]=r[i]+j
    r[i]=r[i]/n

# Постройка гистограммы распределения выборочных средних    
plt.hist(r, normed=True, label='EHyst')
plt.ylabel('number of samples, %')
plt.xlabel('$x$')

# Постройка графика плотности нормального распределения

# для распределения хи-квадрат с k степенями свободы
# теоретическое мат.ожидание - k, а дисперсия - 2k
sgm = np.sqrt(2*k/n)
norm = sts.norm(loc=k, scale=sgm)

x = np.linspace(k-5,k+5,1000)
pdf = norm.pdf(x)
plt.plot(x, pdf, label='Theoretical PDF', color='red')
plt.legend(loc='upper right')
plt.show()
```
