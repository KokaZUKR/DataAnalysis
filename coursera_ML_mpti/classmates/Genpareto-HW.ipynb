```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import genpareto
from scipy.stats import norm
from scipy.stats import rv_continuous
%matplotlib inline
```

# https://ru.wikipedia.org/wiki/Распределение_Парето

#### http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.genpareto.html#scipy.stats.genpareto

##### genpareto.pdf(x, c) = (1 + c * x)**(-1 - 1/c)
зададим распределенную по парето случайную величину, с заданным коэффициентом масштаба

```python
c = 0.1
# сгенерируем 1000 значений
r = genpareto.rvs(c, size=1000)
```

```python
#рисуем график плотности вероятности с гистограммой
fig, ax = plt.subplots(1,1)
x = np.linspace(genpareto.ppf(0.01, c),genpareto.ppf(0.99, c), 100)
ax.plot(x, genpareto.pdf(x, c),'r-', lw=5, alpha=0.6, label='pareto pdf')
plt.title("Pareto PDF, histogram")
plt.xlabel('X')   # подпись оси OX
plt.ylabel('Value')   # подпись оси OY
ax.hist(r, normed = True, histtype='stepfilled', alpha=0.2, label = "histogram")
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
# теоретическое среднее, дисперися, ассиметрия, эксцесс
mean, var, skew, kurt = genpareto.stats(c, moments='mvsk')
```

```python
# генерируем 1000 выборок объема 5 
m = 1000
n = 5
matrix = np.array
matrix = [[0] * n for i in range(m)]
for i in range(m): 
    matrix[i] = genpareto.rvs(c, size=n)
```

```python
# массив выборочных средних
mean_pareto = np.array
mean_pareto = [[0] for i in range(m)]
for i in range(m): mean_pareto[i] = np.mean(matrix[i])
```

```python
#гистограмма для выборочных средних с нормальным распределением
fig, ax = plt.subplots(1,1)
ax.hist(mean_pareto, normed=True, histtype='stepfilled', alpha=0.2, label = "histogram")
ax.legend(loc='best', frameon=False)
plt.xlabel('value mean')   # подпись оси OX
plt.ylabel('frequency')   # подпись оси OY
ax.plot(x, norm.pdf(x, loc = mean, scale = (var/n)**0.5),'r-', lw=6, alpha=0.6, label="norm pdf")  
# в значения нормального распределения подставлены соответвующие значения распределения Парето по Ц.П.Т
plt.show()
```
