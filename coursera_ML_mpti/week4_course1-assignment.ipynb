# Выполнение задания по проверке ЦПТ.

```python
#Импортируем все необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
#Для выполнения задания выбрал распределение Хи-квадрат
#Построим распределение для k = 3
k = 4
rv = sts.chi2(k)
x = np.linspace(0,10,1000)
cdf = rv.cdf(x)
plt.plot(x,cdf)
plt.legend()
plt.title("CDF ($\chi^2_k$)")
```

```python
#Построим теоретическую плотность распределения
pdf = rv.pdf(x)
plt.plot(x, pdf, label="$k=%s$" % k, color = "red")
plt.legend()
plt.title("PDF ($\chi^2_k$)")
```

```python
#Сгенерируем 1000 выборок размером 5 и построим гистограмму
plt.subplots()
means_x = []
for i in range(1000):
    t = rv.rvs(5)
    t_mean = t.mean()
    means_x.append(t_mean)
print("Measured mean is:", np.mean(means_x))
print("Measured std is:", np.std(means_x))
plt.hist(means_x,normed=True)
plt.title("Histogram ($\chi^2_k$), n = 5")
```

```python
#Сгенерируем 1000 выборок размером 10 и построим гистограмму
plt.subplots()
means_x = []
for i in range(1000):
    t = rv.rvs(10)
    t_mean = t.mean()
    means_x.append(t_mean)
print("Measured mean is:", np.mean(means_x))
print("Measured std is:", np.std(means_x))
plt.hist(means_x,normed=True)  
plt.title("Histogram ($\chi^2_k$), n = 10")
```

```python
#Определим стандартное отклонение и среднее Хи-квадрат. Их мы будем использовать для построения нормального распределения
vib_std = rv.std()
vib_mean = rv.mean()
print(rv.std())
print(rv.mean())
```

```python
#Теперь построим те же гистограммы, что и ранее и построим еще дополнительно плотность соответствующего нормального распределения.
plt.subplots()
means_x = []
for i in range(1000):
    t = rv.rvs(5)
    t_mean = t.mean()
    means_x.append(t_mean)
print("Measured mean is:", np.mean(means_x))
print("Measured std is:", np.std(means_x))
print("Theoretical std is:", vib_std**2/5)
plt.hist(means_x,normed=True)
x = np.linspace(0,8,1000)
plt.plot(x,mlab.normpdf(x,vib_mean,vib_std**2/5), color = "r")
plt.title("Histogram ($\chi^2_k$), n = 5")
```

```python
plt.subplots()
means_x = []
for i in range(1000):
    t = rv.rvs(10)
    t_mean = t.mean()
    means_x.append(t_mean)
print("Measured mean is:", np.mean(means_x))
print("Measured std is:", np.std(means_x))
print("Theoretical std is:", vib_std**2/10)
plt.hist(means_x,normed=True)
x = np.linspace(0,8,1000)
plt.plot(x,mlab.normpdf(x,vib_mean,vib_std**2/10), color = "r")
plt.title("Histogram ($\chi^2_k$), n = 10")
```
