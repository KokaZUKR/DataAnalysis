```python
import collections
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
words_bysig
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
anagrams_all
```

```python
words_bylength = collections.defaultdict(list)
```

```python
for word in wordclean:
    words_bylength[len(word)].append(word)
```

```python
words_bylength
```

```python
anagrams_bylength = {}
```

```python
for length, words in words_bylength.items():
    anagrams_bylength[length] = {word: anagram_fast(word) for word in words if len(anagram_fast(word)) > 1}
```

```python
anagrams_bylength
```

```python
for length, words in words_bylength.items():
    anagrams_bylength[length] = { word: len(anagram_fast(word)) - 1 for word in words if len(anagram_fast(word)) > 1}
```

```python
anagrams_bylength
```

```python
anagrams_bylength = {}
for length, words in words_bylength.items():
    anagrams_bylength[length] = sum(len(anagram_fast(word)) - 1 for word in words if len(anagram_fast(word)) > 1)/2
```
