```python
wordclean = [word.strip().lower() for word in open('words','r')]
```

```python
wordclean[:10]
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
wordclean[:10]
```

```python
sorted('lives') 
```

```python
sorted('lives') == sorted('elvis')
```

```python
sorted('hate') == sorted('hated')
```

```python
def anagram(myword):
    return [word for word in wordclean if sorted(word) == sorted(myword)]
```

```python
anagram('aa')
```

```python
anagram('dictionary')
```

```python
%timeit anagram('dictionary')
```
